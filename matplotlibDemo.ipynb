```python
#Paso 1, importar las librerias necesarias
#importar numpy para generar números aleatorios
import numpy as np
import matplotlib.pyplot as plt #para dibujar la gráfica
from matplotlib import style #para darle estilo a la grid de la gráfica
%matplotlib inline #para poder mostrar la gráfica en el notebook de jupyter
```

```python
#Paso 2, definir o importar el data set
#generar los números aleatorios (10 en total)
randomNumber = np.random.rand(10) #esta será nuestro dataset
```

```python
#ver los números generados
print(randomNumber)
```

```python
#Paso 3, establecer los parámetros para generar la gráfica
#establecer el estilo
style.use('ggplot')

#dibujar los números aleatorios
plt.plot(randomNumber,'g', label='línea uno', linewidth=2) #la g indica el color de la línea (g=green)

#darle nombre al eje X (índice)
plt.xlabel('Rango')

#darle nombre al eje Y
plt.ylabel('Números')

#Título de la gráfica
plt.title('Gráfica de números aleatorios')

#si estamos en VS Code, la gráfica se muestra inmediatamente
```

```python
#Paso 4 mostrar la gráfica recien creada, aplica para Jupyter notebook
plt.legend() #generar la gráfica según los parámetros
plt.show() #mostrarla en pantalla
```
