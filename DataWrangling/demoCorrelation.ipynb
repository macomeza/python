## Enunciado
Eres un administrador de una escuela. Algunas escuelas de tu estado tienen un rendimiento debajo del promedio academicamente. Tu jefe esta bajo presión por los padres frustados y los votantes se te acercan para averiguar porque estas escuelas tienen bajo rendimiento.
Ahora, para mejorar el rendimiento de las escuelas, necesitas aprender más sobre estas escuelas y sus estudiantes, así como un negocio necesita comprender sur fortalezas y debilidades y a sus clientes. Los datos incluyen datos demograficos, la facultad de la escuela y variables de ingresos.
Esto se puede lograr mediante tecnicas de analisis exploratorio de datos, incluyendo:
- Leer los datos en un data frame de pandas
- Descibir los datos para encontrar más detalles
- Buscar la correlación entre "almuerzo reducido / reduced_lunch" y "calificación de la escuela / school_rating"

```python
import pandas as pd
df = pd.readf = pd.read_csv('https://raw.githubusercontent.com/LearnDataSci/articles/master/Essential%20Statistics/middle_tn_schools.csv')
df.describe()
```

```python
#fase 2, seleccionar solo la columna del almuerzo y el rating, ver como se relaciona con el rating
df[['reduced_lunch', 'school_rating']].groupby(['school_rating']).describe()
```

```python
#encontrar la correlación
df[['reduced_lunch', 'school_rating']].corr()
```
