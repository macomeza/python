```python
#importar librerias
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#datos ficticios de muerte, las causas más comunes de muerte en un país a tal año
#causas de muertes
causas = 'Enfermedades crónicas','heridas no intencionales','Alzheimers','Influenza y Neumonia','Covid','Otras'

#percentil
percentil = [62,5,4,2,1,26]
```

```python
#establecer las propiedades del pie
#tamaño
plt.figure(figsize=(10,10))

#mostrar aparte al más común
explode = (0.05,0,0,0,0,0.05)

#propiedades
plt.pie(percentil,labels=causas,explode=explode, autopct='%1.1f%%',startangle=45)

#establecer el eje igual para dibujar un circulo
plt.axis('equal')

#título
plt.title('Causas de muerte')

plt.show()
```
