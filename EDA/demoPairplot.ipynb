```python
#importar librerias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#ver las gráficas en el notebook
%matplotlib inline
```

```python
#importar el data set
df_auto_dataset = pd.read_csv('./auto-mpg.csv')
```

```python
#ver los primeros registros
df_auto_dataset.head()
```

```python
#escribir una función para catalogar el origen del vehiculo
#1 - USA, 2 - Europa, 3 - Asia
def origin(num):
    if num==1:
        return 'USA'
    elif num==2:
        return 'Europa'
    else:
        return 'Asia'

#aplicar la función
df_auto_dataset['origin'] = df_auto_dataset['origin'].apply(origin)
```

```python
#ver los primeros para revisar si ya aplicó la función de origen al país
df_auto_dataset.head()
```

```python
#dibujar la gráfica de pareja usando sns para las mpg, peso, origen y matiz por origen, establecer el tamaño a 4
sns.pairplot(df_auto_dataset[['mpg','weight','origin']],hue='origin',size=4,diag_kind="hist")
```
