# Enunciado del caso

Se nos ha dado el dataset que enlista las causas que conducen a muertes en el estado de Ohio del año 2012.

Usando dos listas una de causas de muertes y otra del percentil, dibuje una gráfica de pie que visualice el set de datos.

```python
#importar las librerias
import matplotlib.pyplot as plt

#si usa Jupyter quite el comentario de la siguiente línea
#%matplotlib inline
```

```python
#Datos del estado de Ohio, las causas líderes de muerte en el estado del año 2012-
#causa de las muertes
cause = 'Chronic Diseases','Unintentional Injuries', 'Alzheimers', 'Influenza and Pneumonia', 'Sepsis','Others'
#percentil
percentile = [62,5,4,2,1,26]
```

```python
#establecer las propiedades de la gráfica
#establecer el tamaño de la figura
plt.figure(figsize=(10,10))

#destacar la porción más grande del dataset
explode = (0.05,0,0,0,0,0)

#establecer propiedades del pie
plt.pie(percentile, labels=cause, explode=explode,autopct='%1.1f%%',startangle=70)

#establecer el eje equal para dibujarlo como un círculo
plt.axis('equal')

#darle nombre a la gráfica
plt.title('Estado de Ohio - 2012 : Causas principales de muerte')

#mostrar la gráfica
plt.show()
```
