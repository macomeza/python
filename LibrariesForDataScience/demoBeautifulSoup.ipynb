## Enunciado

Hacerle un scraping al sitio de simplilearn usando la libreria Beautiful Soup y llevar a cabo las siguientes tareas:

- Ver e imprimir el contenido de la pagina de Simplilearn en un formato adecuado
- Ver el encabezado y el título
- Imprimir las referencias de enlaces presentes en la página

```python
#importar las librerias
from bs4 import BeautifulSoup
import requests
```

```python
#ingresar el url para hacerle el scraping
url = 'http://www.guatesuper.com'
```

```python
#accesar a la url con el objeto request
result = requests.get(url)

#hacer el scraping del contenido de la página web
webpage = result.content

#crear el objeto del soup para parsear la página web usando el parser de html
sl_soup = BeautifulSoup(webpage, 'html.parser')
```

```python
#cerrar el objeto de resultado
result.close()
```

```python
#Ver el contenido del objeto
sl_soup.contents
```

```python
#usar el método prettify para ver la salida con un buen formato (con etiquetas html)
print (sl_soup.prettify())
```

```python
#ver el encabezado del objeto Soup
print (sl_soup.head)
```

```python
#ver el título
sl_soup.title
```

```python
#buscar todos los enlaces presentes en la pagina
for href in sl_soup.findAll('a', href=True):
    print (href['href'])
```
