```python
#importar el dataset de boston desde scikit
from sklearn.datasets import load_boston
#importar matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
#cargar los datos de boston
boston_real_state_data = load_boston()
```

```python
#ver el dataset
print(boston_real_state_data)
```

```python
#definir el eje x para los datos
x_axis = boston_real_state_data.data
```

```python
#definir el eje y para el objetivo
y_axis = boston_real_state_data.target
```

```python
#mostrar un histograma
style.use('ggplot')
plt.figure(figsize=(7,7))
plt.hist(y_axis, bins=50)
plt.xlabel('precio en 1000s US$')
plt.ylabel('número de casas')
plt.show()
```

```python
#graficar una de dispersión
style.use('ggplot')
plt.figure(figsize=(7,7))
plt.scatter(boston_real_state_data.data[:,5], boston_real_state_data.target)
plt.ylabel('precio en 1000s')
plt.xlabel('número de casas')
```

```python
#importar libreras
import seaborn as sns
```

```python
#cargar los datos de vuelos
flight_data = sns.load_dataset('flights')
```

```python
flight_data.head()
```

```python
#usar el método pivot para girar el dataset
flight_data = flight_data.pivot('month','year','passengers')
```

```python
#ver el dataset
flight_data
```

```python
#generar el mapa de calor
sns.heatmap(flight_data)
```

```python
#creremos un dataset con porcentajes y etiquetas
job_data = ['40','20','17','8','5','10']
labels = 'IT','Finanzas','Mercadeo','Admon','RRHH','Operaciones'

#explotar la primera porción
explode = (0.1,0,0,0,0,0)

#dibujar el gráfico de pie y establecer los parámetros
plt.pie(job_data,labels=labels,explode=explode)

#mostrar la gráfica
plt.show()
```
